```python
file_name = 'test.csv'
f = open(file_name, 'r')
rdr = csv.reader(f)
pm = []
time = []
for line in rdr:
    pm.append(line[1])
    time.append(line[3])
f.close()
```

```python
pm = [int(x) for x in pm[1:]]
```

```python
chk = {}
for i in range(len(pm)):
    if pm[i] < 30:
        chk[time[i]] = pm[i]
```

```python
s = '2021-03-17 06:50'
e   = '2021-03-17 07:10'
count = 0
start = 0
end = 0
for t in time[1:]:
    count = count + 1
    if s in t:
        print('start : ', count)
        start = count
    if e in t:
        print('end :   ', count)
        end = count
```

```python
f = open('flf.csv', 'r')
rdr = csv.reader(f)
num = 0
b = []
for line in rdr:
    num = num + 1
    if len(line[0].split(',')) != 40:
        print(num, len(line[0].split(',')))
    else:
        b.append(line[0].split(','))    
f.close()
```

```python
f = open('last_lf2.csv','w', newline='')
writer = csv.writer(f)
writer.writerow(["type %d"%i for i in range(len(b))])
for i in range(len(b[0])):
    writer.writerow([b[j][i] for j in range(len(b))])
f.close()
```
