```python
data = pd.read_csv("public_data.csv", sep=",",engine='python',encoding='CP949')
```

```python
plt.figure(figsize=(16,9))
data['pm10'].plot()
plt.show()
```

```python
X = np.array(data['pm10'])
y = np.array(data['pm10'][3:])
```

```python
X.shape
```

```python
y.shape
```

```python
feature_list = []
label_list = []

window_size = 3
for i in range(len(X)-window_size):
    feature_list.append(np.array(data['pm10'].iloc[i:i+window_size]))
    label_list.append(np.array(data['pm10'].iloc[i+window_size]))
```

```python
X = np.array(feature_list)
y = np.array(label_list)

X.shape[1]
# X_test = np.array(X_test)
# X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

X = np.reshape(X, (X.shape[0],X.shape[1],1))
```

```python
print(X.shape)
print(y.shape)
```

```python
#y.shape
X.shape[1]
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import EarlyStopping, ModelCheckpoint
from keras.layers import LSTM
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D, LSTM
```

```python
model = Sequential()

model.add(LSTM(units = 50, activation='relu', input_shape =(X.shape[1],1)))
model.add(Dropout(0.2))

model.add(Dense(units=1))
model.compile(optimizer='adam', loss ='mean_squared_error')
```

```python
model.summary()
```

```python
history = model.fit(X, y, epochs=200, validation_split=0.3,batch_size=32, shuffle=True)
```
