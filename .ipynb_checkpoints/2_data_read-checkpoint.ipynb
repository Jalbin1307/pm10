```python
import matplotlib.pyplot as plt
import tensorflow as tf
import csv
import numpy as np
```

```python
file_name = 'test.csv'
f = open(file_name, 'r')
rdr = csv.reader(f)
pm = []
time = []
for line in rdr:
    pm.append(line[1])
    time.append(line[3])
f.close()
```

```python
pm = [int(x) for x in pm[1:]]
```

```python
chk = {}
for i in range(len(pm)):
    if pm[i] > 70:
        chk[time[i]] = pm[i]
```

```python
chk
```

```python
s = '2021-03-11 16:22'
e   = '2021-03-11 16:42'
count = 0
start = 0
end = 0
for t in time[1:]:
    count = count + 1
    if s in t:
        print('start : ', count)
        start = count
    if e in t:
        print('end :   ', count)
        end = count
```

```python
plt.plot(pm[start:end])

plt.show()
```
