```python
import numpy as np
from ast import literal_eval
import csv
```

```python
def save_to_file(num):
    file = open("test.csv", mode="w", newline='')
    writer = csv.writer(file)
    writer.writerow(["PM2.5", "PM10", "PM1.0", "DATE", "ID"])
    for i in num:
        writer.writerow(list(i.values()))
    return
```

```python
f = open("0331.txt",'r')
pm = []
for i in range(0, 18122):
    try :
        line = f.readline()
        date = line[-49:-30]
        id = line[-4:-3]
        data = literal_eval(line[13:-55])
        data['raw'][0]['data']['date'] = date
        data['raw'][0]['data']['id'] = id
        
        pm.append(data['raw'][0]['data'])
    except:
        pass
```

```python
pm
```

```python
save_to_file(pm)
```
